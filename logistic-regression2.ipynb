```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import numpy as np
```

```python
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
# Features (X): [age, income]
X = np.array([
    [30, 40000],
    [40, 60000],
    [50, 80000],
    [20, 20000],
    [35, 50000],
    [45, 70000]
])

# Labels (y): 0 = No, 1 = Yes
y = np.array([0, 1, 1, 0, 0, 1])
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(
        solver='saga',          # REQUIRED
        penalty='elasticnet',   # REQUIRED
        max_iter=2000
    ))
])
```

```python
param_grid = {
    'logreg__C': [0.01, 0.1, 1, 10],
    'logreg__l1_ratio': [0.0, 1.0],   # 0 = L2, 1 = L1
    'logreg__penalty': ['elasticnet'] # Explicitly set this to avoid the warning
}
```

Run cross-validation (GridSearchCV)

```python
grid = GridSearchCV(
    pipeline,
    param_grid,
    cv=3,
    scoring='accuracy',
    #error_score='raise'  # helps debugging
)

grid.fit(X, y)
```

Best model chosen by cross-validation

```python
print("Best parameters:", grid.best_params_)
print("Best cross-validation accuracy:", grid.best_score_)
```

Use the best model for prediction

```python
best_model = grid.best_estimator_

new_customer = np.array([[37, -100000]])
prediction = best_model.predict(new_customer)

print("Prediction:", prediction)  # 0 = No, 1 = Yes
```
